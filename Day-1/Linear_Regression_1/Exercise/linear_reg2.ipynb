```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model 
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
plt.xlabel("year")
plt.ylabel("GDP")
plt.scatter(df.year,df.pci,c="red")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.pci)
```

```python
new_data = pd.DataFrame({'year': [2050]})
predicted_price = reg.predict(new_data)
print(predicted_price)
```

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income (in USD)")
plt.scatter(df.year,df.pci,c="red",marker="+")
plt.plot(df.year,reg.predict(df[['year']]),c="blue")
```
